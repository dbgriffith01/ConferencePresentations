## Exploring the Black-Box

### Discuss the application of Keener and Markov Chain methods (NFL 2013)

<br/>
![Image of Yaktocat](MarkovChain.jpg)
<br/>

Keener's matrix shows the $A_{ij}$ entries for the Cards, Niners, Seahawks, and Rams:

<br/>

$ \quad \quad \quad \mathbf{A} = \left[
  \begin{array}{rrrr}
0 & 0.422680  & 0.470588 & 0.591398  \\
0.577320 & 0 & 0.363636 & 0.702381 \\
0.529412 &  0.636364 & 0 & 0.688525 \\
0.408602 & 0.297619 & 0.311475 & 0
\end{array}
\right] $

<br/>

### Keener's Algorithm

The algorithm requires:

1. Setting up a matrix **A**, where each entry $A_{ij}$ is interpreted as the strength of team $i$ when playing team $j$.
2. These entries can be defined based on anything: a) wins, b) total yards scored, c) *total points scored*.
3. Specifically, we define $A_{ij} = (S_{ij} + 1)/(S_{ij} + S_{ji} + 2)$.
4. The largest eigenvalue, along with the corresponding eigenvector of **A** is the ratings vector, **v**.

#### Interpretation of Keener's first eigenvector

* Since patterns in data can be hard to find in high-dimensional spaces, eigenanalysis is a powerful tool for analyzing data.
* Eigenanalysis will identify the patterns in the data, and express it in simplified form.
* The first eigenvector, corresponding to the largest positive-real eigenvalue, is the ratings vector. 
* For the NFC West 2013 divisional games, the matrix belows gives us $\mathbf{v} = (0.25006, 0.26491, 0.29667, 0.18836)^T$ corresponding to the ratings for Arizona, San Francisco, Seattle, and St. Louis.

### Markov Chain Approach

Markov chains are used to describe a **stochastic process** that moves from one state to another.

* Markov chains have many different applications in Chemistry, Queueing Theory, and as most recently shown by Google in their PageRank algorithm. 
* A Markov chain describes how a system moves from one **state** to another with a specific probability. 
* In our case, let's picture the state of an NFL team that has the "loser trophy".  No team would like to keep it, so you must face other teams and beat them to be able to pass along the trophy.  
![Image of Yaktocat](LoserTrophy.jpg)

* Strong teams are less likely to get the loser trophy from a weaker team, but it is not impossible. 
* The long term (steady-state) probabilities represent the likelihood (or proportion of time) a specific team would hold the loser trophy.
* Teams with higher probabilities represent the weaker teams, while teams with low probabilities represent the stronger teams. 



The matrix of transition probabilities $\mathbf{P}$ can be obtained from the matrix $\mathbf{A}$ shown before for the Cardinals, 49ers, Seahawks, and Rams:

<br/>

$ \quad \quad \quad \mathbf{P} = \left[
  \begin{array}{rrrr}
0 & 0.284697  & 0.316966 & 0.398337  \\
0.351309 & 0 & 0.221279 & 0.427411 \\
0.285505 &  0.343183 & 0 & 0.371312 \\
0.401497 & 0.292444 & 0.306059 & 0
\end{array}
\right] $

<br/>

* This produces the vector of "reversed" ratings: $\mathbf{v}=(0.2597989, 0.2334048, 0.2213564, 0.2854398)^T$.
* Showing the ranking is: Seattle, San Francisco, Arizona, and St. Louis.
